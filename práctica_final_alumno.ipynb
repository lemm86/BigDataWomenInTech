```python
import pandas as pd
```

Cargar datos

```python
# Importamos el dataset de los datos de los pokemon
df = pd.read_csv('dataset_final.csv')
df.head()
```

```python
'''
1. ¿Cuántos pokémon hay en el dataset?
'''
n_pokemons = len(df.axes[0])
print(n_pokemons)
```

```python
'''
2. ¿Cuántos pokémon hay de tipo Poison? (Type 1)
'''
n_poison_type1= df[df['Type 1'] == "Poison"]
n_poison_pokemons = len(n_poison_type1.axes[0])
print(n_poison_pokemons)
```

```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
type_pokemon = df.groupby('Type 1')['Name'].count()
n_type_pokemons = len(type_pokemon.axes[0])
print(n_type_pokemons)
```

```python
'''
2. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''
type_pokemon = df.groupby('Type 1')['Name'].count()
print(type_pokemon)

'''
2.1 ¿cual es el tipo de pokémon con más pokémon? (Type 1)
'''
max_pokemon = df.groupby('Type 1')['Name'].count().max()
print(max_pokemon)
```

```python
'''
3. ¿Cuál es el pokémon más rápido?
'''
faster_pokemon = df['Name'][df['Speed'].max()]
speed_faster_pokemon = df['Speed'].max()
print(faster_pokemon)
print(speed_faster_pokemon)
```

```python
'''
4. ¿Cuántos pokemon tiene una defensa superior a 100?
'''
defense_higher_100 = df[df['Defense']>100]
n_defense_higher_100 = len(defense_higher_100)
print(n_defense_higher_100)
```

```python
'''
5. ¿Cuántos pokemon tiene una defensa superior a 100 y una velocidad superior a 100?
'''
defense_and_speed_higher_100 = defense_higher_100[defense_higher_100['Speed']>100]
n_defense_and_speed_higher_100 = len(defense_and_speed_higher_100)
print(n_defense_and_speed_higher_100)
```

```python
'''
8. Crea un nuevo dataset con los pokemon Legendary
'''
legendary_pokemon = df[df["Legendary"] == True]
no_legendary_pokemon = df[df["Legendary"] == False]
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''
hp_legen = legendary_pokemon['HP'].mean()
attack_legen = legendary_pokemon['Attack'].mean()
defense_legen = legendary_pokemon['Defense'].mean()
sp_atk_legen = legendary_pokemon['Sp. Atk'].mean()
sp_def_legen = legendary_pokemon['Sp. Def'].mean()
speed_legen = legendary_pokemon['Speed'].mean()

hp_no_legen = no_legendary_pokemon['HP'].mean()
attack_no_legen = no_legendary_pokemon['Attack'].mean()
defense_no_legen = no_legendary_pokemon['Defense'].mean()
sp_atk_no_legen = no_legendary_pokemon['Sp. Atk'].mean()
sp_def_no_legen = no_legendary_pokemon['Sp. Def'].mean()
speed_no_legen = no_legendary_pokemon['Speed'].mean()

```

```python
'''
11. Añade una nueva colunma ['Doble tipo'] al dataframe inicial que inndique si el pokemo tiene dos tipos o no
'''

df.loc[df['Type 2'].isnull(), 'Doble_type'] = 'One_type'
df.loc[(df['Type 2'].isnull()) == False, 'Doble_type'] = 'Doble_type'
df
```

```python
'''
10. Crea un nuevo dataframe con el resultado del anterior ejercicio comparando ambos promedios

ejemplo:
                HP  Attack  Defense  Sp. Atk  Sp. Def  Speed
Legendary       99      90       89       91       94     90
No Legendary    80      95       75       12       43     87

'''

df_means = pd.DataFrame()
df_means[""] = ["Legendary", "No Legendary"]
df_means["HP"] = [hp_legen, hp_no_legen]
df_means["Attack"] = [attack_legen, attack_no_legen]
df_means["Defense"] = [defense_legen, defense_no_legen]
df_means["Sp. Atk"] = [sp_atk_legen, sp_atk_no_legen]
df_means["Sp. Def"] = [sp_def_legen, sp_def_no_legen]
df_means["Speed"] = [speed_legen, speed_no_legen]
df_means
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra

precios_compra = map(lambda producto: articulos[producto], compra)
total_compra = reduce(lambda x, y: x + y, precios_compra)

print(total_compra)

# B. Una función que calcule el precio total de la compra con un descuento del 10%

precios_compra = map(lambda producto: articulos[producto], compra)
total_compra = reduce(lambda x, y: (x + y), precios_compra)
total_pagar = total_compra * 0.9

print(total_pagar)

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

precios_compra = map(lambda producto: articulos[producto], compra)
total_compra = reduce(lambda x, y: (x + y), precios_compra)
def total_pagar(total_compra):
    total_pagar = 0
    if total_compra > 100: 
        total_pagar = total_compra * 0.9
    else: 
        total_pagar = total_compra
    return total_pagar
        
print(total_pagar(total_compra))

# D. Una función que calcule el precio total aplicando el IVA (21%)

precios_compra = map(lambda producto: articulos[producto], compra)
total_compra = reduce(lambda x, y: (x + y), precios_compra)
total_con_IVA = total_compra * 0.79

print(total_con_IVA)

# E.lista los artículos cuyo precio es superior a 20€

producto_precio_mayor_20 = list(filter(lambda producto: articulos[producto] > 20, articulos))
print(producto_precio_mayor_20)

```
